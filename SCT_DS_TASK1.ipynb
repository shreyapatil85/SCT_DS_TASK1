```python
import matplotlib.pyplot as plt

# Sample age data
ages = [
    5, 8, 12, 15, 18, 22, 25, 28, 30, 32, 35, 38,
    40, 42, 45, 48, 50, 52, 55, 58, 60, 62, 65, 70, 75, 80
]

# Create histogram
plt.figure()
plt.hist(ages, bins=6)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Histogram of Age Distribution")

plt.show()
```
